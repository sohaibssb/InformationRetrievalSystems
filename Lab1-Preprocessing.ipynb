Задание 2 (2 балла)

Токенизация Mystem vs razdel.tokenize

Токенизируйте текст с помощью razdel и с помощью Mystem. Найдите различия в токенизациях. Что по вашему работает лучше на приведенном тексте?

```python
from razdel import tokenize
from pymystem3 import Mystem

text = """
Вторым и третьим открытыми белыми карликами стали Сириус B и Процион B. В 1844 году директор Кёнигсбергской обсерватории Фридрих Бессель, анализируя данные наблюдений, которые велись с 1755 года, обнаружил, что Сириус, ярчайшая звезда земного неба, и Процион периодически, хотя и весьма слабо, отклоняются от прямолинейной траектории движения по небесной сфере[5]. Бессель пришёл к выводу, что у каждой из них должен быть близкий спутник. Сообщение было встречено скептически, поскольку слабый спутник оставался ненаблюдаемым, а его масса должна была быть достаточно велика — сравнимой с массой Сириуса и Проциона, соответственно.

В январе 1862 года Элвин Грэхэм Кларк, юстируя 18-дюймовый рефрактор, самый большой на то время телескоп в мире (Dearborn Telescope), впоследствии поставленный семейной фирмой Кларков в обсерваторию Чикагского университета, обнаружил в непосредственной близости от Сириуса тусклую звёздочку. Это был спутник Сириуса, Сириус B, предсказанный Бесселем[6]. А в 1896 году американский астроном Д. М. Шеберле открыл Процион B, подтвердив тем самым и второе предсказание Бесселя.

В 1915 году американский астроном Уолтер Сидней Адамс измерил спектр Сириуса B. Из измерений следовало, что его температура не ниже, чем у Сириуса A (по современным данным, температура поверхности Сириуса B составляет 25 000 K, а Сириуса A — 10 000 К), что, с учётом его в 10 000 раз меньшей, чем у Сириуса A, светимости указывает на очень малый радиус и, соответственно, высокую плотность — 106 г/см3 (плотность Сириуса ~0,25 г/см3, плотность Солнца ~1,4 г/см3).
"""

tokenized_text = tokenize(text)
razdel_tokens = []

for token in tokenized_text: 
    token_text = token.text
    razdel_tokens.append(token_text)


#/////////////////////////////////////////
    
mystem = Mystem()
mystem_tokens = mystem.lemmatize(text)
mystem_tokens = [token for token in mystem_tokens if token != ' ' and token != '\n']

#/////////////////////////////////////////
          

print("Различия в токенизациях:")
print("Razdel:", razdel_tokens)
print("Mystem:", mystem_tokens)

```

Задание 2 - Заметки

Токенизация Razdel:

Успешно разбивает текст на значимые лексемы.
Сохраняет знаки препинания и хорошо обрабатывает сложные структуры, такие как числа и квадратные скобки.

Токенизация Mystem:

Предоставляет лемматизированные лексемы, преобразуя слова в их базовые формы.
По-другому обрабатывает знаки препинания, отделяя их от слов.
Сохраняет пробелы между словами и знаками препинания.

Лемматизация Mystem vs Pymorphy

Лемматизируйте текст с помощью mystem и pymorphy. Найдите различия в лемматизации. Что по вашему работает лучше на приведенном тексте?

Важно: для пайморфи используйте токенизацию из mystem, чтобы исключить влияние токенизации на результат. Анализируйте только значимые различия, а не технические особенности (не сравнивайте скорость работы и удобность интерфейса).


```python
from pymorphy2 import MorphAnalyzer
from pymystem3 import Mystem

mystem = Mystem()
mystem_tokens = mystem.analyze(text)
mystem_lemmas = [token["analysis"][0]["lex"] for token in mystem_tokens if "analysis" in token and token["analysis"]]

#////////////////////////////////////////

pymorphy = MorphAnalyzer()
pymorphy_lemmas = [pymorphy.parse(token)[0].normal_form for token in mystem_lemmas]

#////////////////////////////////////////

print("Mystem Lemmas:", mystem_lemmas)
print("Pymorphy Lemmas:", pymorphy_lemmas)
```

Дизамбигуация в Mystem
*необязательное задание на 10 баллов

Одна из фичей майстема - контекстная дизамбигуация (т.е. определение правильного разбора в зависимости от контекста).

Запустите mystem на очищенном тексте с дизамбигуацией и без и проверьте есть ли какие-то различия.

Run mystem on cleaned text with and without disambiguation and check if there are any differences.

```python
from pymystem3 import Mystem

cleaned_text = """
Второй и третий открытый белый карлик становятся Сириус B и Процион B. В 1844 году директор Кенигсбергской обсерватории Фридрих Бессель, анализируя данные наблюдений, которые велись с 1755 года, обнаружил, что Сириус, яркая звезда земного неба, и Процион периодически, хотя и весьма слабо, отклоняются от прямолинейной траектории движения по небесной сфере. Бессель пришел к выводу, что у каждой из них должен быть близкий спутник. Это сообщение было встречено скептически, поскольку слабый спутник оставался ненаблюдаемым, а его масса должна была быть достаточно большой — сравнимой с массой Сириуса и Проциона соответственно.
В январе 1862 года Элвин Грэхэм Кларк, юстировав 18-дюймовый рефрактор, самый большой на тот момент телескоп в мире (Dearborn Telescope), впоследствии поставленный семейной фирмой Кларк в обсерваторию Чикагского университета, обнаружил в непосредственной близости от Сириуса тусклую звездочку. Это был спутник Сириуса, Сириус B, предсказанный Бесселем. А в 1896 году американский астроном Д. М. Шеберле открыл Процион B, подтвердив тем самым и второе предсказание Бесселя.
В 1915 году американский астроном Уолтер Сидней Адамс измерил спектр Сириус B. Из измерения следовало, что его температура не ниже, чем у Сириуса A (по современным данным, температура поверхности Сириус B составляет 25 000 K, а Сириуса A — 10 000 K), что, с учетом его в 10 000 раз меньшей, чем у Сириуса A, светимости, указывает на очень малый радиус и, соответственно, высокую плотность — 106 г/см3 (плотность Сириуса ~0.25 г/см3, плотность Солнца ~1.4 г/см3).
"""

mystem = Mystem()

result_no_disambiguation = mystem.analyze(cleaned_text)
lemmas_no_disambiguation = [token["analysis"][0]["lex"] for token in result_no_disambiguation if "analysis" in token and token["analysis"]]

result_disambiguation = mystem.analyze(cleaned_text)
lemmas_disambiguation = [token["analysis"][0]["lex"] for token in result_disambiguation if "analysis" in token and token["analysis"]]

print("Lemmas without disambiguation:", lemmas_no_disambiguation)
print("Lemmas with disambiguation:", lemmas_disambiguation)

```
